# MCU  Movie data scraping from wikip


```python
#importing libraries 

import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url="https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films"
table1_class="wikitable plainrowheaders"

response=requests.get(url)
soup=BeautifulSoup(response.text,"html.parser")
```

```python
movies=soup.find_all("table",attrs={"class":table1_class})
```

```python
movies
```

```python
movies_table=pd.read_html(str(movies))
```
